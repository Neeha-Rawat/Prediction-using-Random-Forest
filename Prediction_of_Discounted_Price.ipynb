```python
def clean(data): #function to clean the datasets
  title=data["title"].tolist() #changing title column to a list to apply for loop
  for i in title : #for each value in title
    if "Men" in i: #if Men is present in the title
      data["maincateg"].fillna("Men",inplace=True) #fill the null value in maincateg as Men
    elif "Women" in i: #if Women is present in the title
      data["maincateg"].fillna("Women",inplace=True) #fill the null value in maincateg as Women
  
  cols = ["Rating","norating1", "noreviews1", "star_5f", "star_4f","star_3f","star_1f"] #all the columns having null values 
  for col in cols: #for each column having null value
      data[col].fillna(0, inplace=True) #fill the null value with 0

  data=data.drop(['id','title'],axis=1)#dropping id and title column
  return data#returning clean data
```

```python
data_train=data_train.drop(['Offer %'],axis=1)#dropping Offer % column
```

```python
train=clean(data_train)#applying clean function to training data
test=clean(data_test)#applying clean function to test data
```

```python
train.head() #showing top 5 rows of cleaned data
```

```python
from sklearn import preprocessing #to pre-process before modelling 
le= preprocessing.LabelEncoder()#using label encoder i.e. giving specific integer labels to specific values

col=['maincateg','platform']#column to apply label encoding for maincateg giving value 0 for Men and 1 for Women, in platform encoding amazon as 0 and flipkart as 1
for i in col: #for each column 
  train[i]=le.fit_transform(train[i])#apply label encoding in training dataset
  test[i]=le.transform(test[i])#apply label encoding in test dataset
  print(le.classes_)#printing the classes i.e. the unique values for which label encoding is applied
    
train.head()#showing top 5 rows of preprocessed dataset
```

```python
y=train["price1"] #fixing target variable
X=train.drop(["price1"],axis=1) #separating target variable from dataset
```

```python
from sklearn.model_selection import train_test_split #to split the dataset into training and test dataset to check accuracy

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression 
import xgboost
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

```python
def score(model, title = "Default"):#a function to measure accuracy of the models
    model.fit(X_train, y_train)#fitting our model
    preds = model.predict(X_test)#predicting y_test
    df=pd.DataFrame({'Actual':y_test,'Predicted':preds})#comparing actual and predicted value of y_test
    df
    print ('Mean Absolute Error :',metrics.mean_absolute_error(y_test,preds))#finding mae
    print ('Mean Squared Error :',metrics.mean_squared_error(y_test,preds))#find mse
    accuracy=model.score(X_test,y_test)#to check accuracy score
    print("Accuracy",accuracy*100,"%")
```

```python
lr=LinearRegression()
score(lr,"Linear Regression")
```

```python
xgb=xgboost.XGBRegressor(n_estimators=1000)
score(xgb,"XGB Regressor")
```

```python
rf=RandomForestRegressor()
score(rf,"Random Forest Regressor")
```

```python
X_tr=X #storing training data other than target variable
y_tr=y #storing target variable from training data
X_te=test #storing given test data 
```

```python
pred=RandomForestRegressor(n_estimators=1000)#applying Random Forest Regression
pred.fit(X_tr,y_tr)#fitting the model
predictions=pred.predict(X_te)#predicting discounted prices for the given test dataset
```
